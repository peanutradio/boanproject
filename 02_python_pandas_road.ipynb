```python
#데이터초기화 작업
data = pd.read_csv(io.BytesIO(road_file['road_data.csv']))
data.head()
data2 = data.copy()
data.head()
```

```python
#원하는 합계를 결과 데이터에 포함
data2['중상자수이상'] = data2['사망자수'] + data2['중상자수']
data2['사고대비사망자'] = data2['사망자수'] / data2['사고건수'] * 100
data2
```

```python
#데이터를 다른 형태로 저장
data2.to_csv("data2.csv")
data2.to_excel("data2.xlsx")
data2.to_json('data2.json')
data2.to_html('data2.html')
```

```python
data2.index = ['2018년','2019년','2020년','2021년','2022년']
data2
```
